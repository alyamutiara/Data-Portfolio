### Soal 3
Sebagai BI Analyst PT Sejahtera Bersama, kita akan membuat sebuah table master yang berisikan informasi :
- CustomerEmail (cust_email)
- CustomerCity (cust_city)
- OrderDate (order_date)
- OrderQty (order_qty)
- ProductName (product_name)
- ProductPrice (product_price)
- ProductCategoryName (category_name)
- TotalSales (total_sales)
Urutkan data tersebut berdasarkan tanggal transaksi yang paling awal sampai yang paling akhir.

### BigQuery

```python
SELECT 
  o.date AS order_date, 
  c.CategoryName AS category_name,
  p.ProdName AS product_name,
  p.Price AS product_price,
  o.quantity AS order_qty,
  p.Price * o.Quantity AS total_sales,
  cus.CustomerEmail AS customer_email,
  cus.CustomerCity AS customer_city
FROM `emerald-skill-352503.vix_sales_report.orders` AS o
LEFT JOIN `emerald-skill-352503.vix_sales_report.product_category` AS c
ON LEFT(o.ProdNumber, 2) = c.CategoryAbbreviation
LEFT JOIN `emerald-skill-352503.vix_sales_report.products` AS p
ON o.ProdNumber = p.ProdNumber
LEFT JOIN `emerald-skill-352503.vix_sales_report.customers` AS cus
ON o.CustomerID = cus.CustomerID
LIMIT 3
```
